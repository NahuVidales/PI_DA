```python
#importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
courseraCourses = pd.read_csv('Coursera_courses.csv')
edxCourses = pd.read_csv('edx_courses.csv')
udemyCourses = pd.read_csv('udemy_courses.csv')
```

```python
#Eliminamos las filas duplicadas de todos
courseraCourses = courseraCourses.drop_duplicates()
edxCourses = edxCourses.drop_duplicates()
udemyCourses = udemyCourses.drop_duplicates()
```

```python
#creamos la columna institucion que recorrera la columna de 'institution' y si encuentra la palabra 'univer'
#escribira universidad en la columna institucion y en caso que encuentre la palabra 'coleg' escribira colegio
courseraCourses['tipoInstitucion'] = courseraCourses['institution'].apply(lambda x: 'universidad' if 'univer' in x.lower() else 'colegio' if 'coleg'  in x.lower() else 'empresa y otras instituciones')
courseraCourses
```

```python
#Haremos lo mismo con el dataframe edx
edxCourses['tipoInstitucion'] = edxCourses['institution'].apply(lambda x: 'universidad' if 'univer' in x.lower() else 'colegio' if 'coleg'  in x.lower() else 'empresa y otras instituciones')
```

```python
#Guardamos los valores de la columna 'num_suscribers' en una nueva columna si la columna de 'is_paid' es igual a 'True'
udemyCourses['num_suscribers_paid'] = udemyCourses.apply(lambda row: row['num_subscribers'] if row['is_paid'] else 0, axis=1)
udemyCourses['num_suscribers_Nopaid'] = udemyCourses.apply(lambda row: row['num_subscribers'] if not row['is_paid'] else 0, axis=1)
udemyCourses
```

```python
#vamos a sumar todos los valores de la columna price de udemy multiplicado por el numero de alumnos inscritos
#para ver cuanto dinero se ha ganado con los cursos
udemyCourses['price'] = udemyCourses['price'].astype(int)
udemyCourses['Ganado'] = udemyCourses['price'] * udemyCourses['num_subscribers']
udemyCourses['Ganado'].sum()
```

```python
courseraCourses.to_csv('Coursera_coursesWO.csv', index=False)
edxCourses.to_csv('edx_coursesWO.csv', index=False)
udemyCourses.to_csv('udemy_coursesWO.csv', index=False)
```
